```python
from pandas import read_excel
from requests import get
from bs4 import BeautifulSoup
```

```python
df = read_excel("Output Data Structure.xlsx")
```

```python
list_specific_tag_not_found = []

for i in df["URL"] :
    r = get(i)
    if "td-post-content tagdiv-type" not in r.content.__str__() :
        list_specific_tag_not_found.append(i)
```

```python
lst_title = []
lst_exceptions_article = []

def make_write_file(file_title : str , file_content : str | None) :
    with open(file_title,'w',encoding='utf-16') as file :
        file.write(file_content)

for x in range(df.__len__()) :
    
    i = df.loc[x,"URL"]
    j = df.loc[x,"URL_ID"]
    try :
        r = get(i)

        soup = BeautifulSoup(r.content , "html.parser")

        z = soup.find("div",class_="td-post-content tagdiv-type")
        if type(z) == type(None) :
            z = soup.find("div",class_="tdb-block-inner td-fix-index")
            if type(z) == type(None) :
                lst_title.append(None)
                make_write_file(str(j)+".txt" , None)
                continue

        lst_title.append(soup.title.text)
        make_write_file(str(j)+".txt" , z.text)
    except UnicodeEncodeError :
        lst_exceptions_article.append("Unicode Error at URL ID : "+str(j))
    except Exception as e :
        lst_exceptions_article.append("Exception at URL ID: "+str(j))
    
```

```python
lst_exceptions_article
```

```python
df["Title"] = lst_title
```
